```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# 1) प्रशिक्षणासाठी डेटा (Training Data)
X = np.array([1, 2, 3, 4, 5], dtype=float)   # इनपुट डेटा
Y = np.array([2, 4, 6, 8, 10], dtype=float)  # आउटपुट डेटा

# 2) साधं मॉडेल तयार करणे
model = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])   # एक लेयर, एक न्यूरॉन
])

# 3) मॉडेल compile करणे
model.compile(optimizer='sgd', loss='mean_squared_error')

# 4) मॉडेल ट्रेन करणे
model.fit(X, Y, epochs=500, verbose=0)

# 5) इनपुट देऊन आउटपुट बघणे
inp = float(input("कृपया एक नंबर टाका: "))   # यूजरकडून इनपुट घेणे
prediction = model.predict(np.array([inp]))   # numpy array मध्ये बदलणे
print(f"आपण दिलेला नंबर: {inp}")
print(f"TensorFlow चा अंदाज: {prediction[0][0]}")
```
