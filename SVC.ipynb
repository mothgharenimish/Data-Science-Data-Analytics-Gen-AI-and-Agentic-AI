```python
##### import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

data = {

    'Age' : [12,30,90,23,123,90,87,34,44,90],
    'Income' : [11000,9000,8900,10000,89098,1110,20000,909000,10002,120],
    'Buy' : [1,1,1,0,0,1,1,0,0,1]
}

df = pd.DataFrame(data)

df

x = df[['Age','Income']]
y = df[['Buy']]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = SVC(kernel='linear')

model.fit(X_train, y_train)

# Step 6: User Input घ्या
age = int(input("Enter Customer Age: "))
income = int(input("Enter Customer Income: "))

# Step 7: Prediction करा
prediction = model.predict([[age, income]])

if prediction[0] == 1:
    print("✅ Customer will BUY the product.")
else:
    print("❌ Customer will NOT buy the product.") 
```
