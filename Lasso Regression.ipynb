```python
import numpy as np
from sklearn.linear_model import Lasso
```

```python
X = np.array([

    [700,3,28,19],
    [900,4,56,33],
    [550,2,44,32],
    [854,3,55,29]
])

Y = np.array([99,95,60,88])

model = Lasso(alpha=1.0)
print(model)
model.fit(X,Y)

print("Weights:", model.coef_)
print("Bias:", model.intercept_)
```
